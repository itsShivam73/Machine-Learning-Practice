```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/House_Rent_Dataset.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=['Posted On' , 'Floor','Area Locality'])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train, y_test = train_test_split(df.drop(columns=['Rent']) , df['Rent'],test_size=0.2 , random_state=0)
```

```python
x_train.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder , OrdinalEncoder
transformer = ColumnTransformer(transformers=(
    ('trnf1' , OneHotEncoder(sparse_output=False ,drop='first'),['City' , 'Tenant Preferred','Point of Contact']),
    ('trnf2' , OrdinalEncoder(categories=[
                             ['Built Area' ,'Carpet Area','Super Area'],
                             ['Unfurnished','Semi-Furnished','Furnished']]),
     ['Area Type' ,'Furnishing Status']
     )
) ,remainder= 'passthrough'
                                )
```

```python
transformer
```

```python
x_train_new = transformer.fit_transform(x_train)
```

```python
x_test_new = transformer.transform(x_test)
```

```python
from sklearn.pipeline import Pipeline , make_pipeline
```

```python
pipe = Pipeline([

                    ('trfn' , transformer)
])
```

```python
pipe.fit(x_train,y_train)
```

```python
pipe.named_steps
```
