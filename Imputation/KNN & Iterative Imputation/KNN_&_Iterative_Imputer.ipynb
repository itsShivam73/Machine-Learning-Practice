Import library & datasets

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/titanic_datasets.csv',usecols=['Age','Pclass',"Fare",'Survived'])
df.head()
```

```python
df.isnull().mean()*100
```

Train_Test_splits

```python
x = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test ,y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.head(3)
```

KNN Imputer

```python
from sklearn.impute import KNNImputer
```

```python
knn = KNNImputer()
x_train_trf = knn.fit_transform(x_train)
x_test_trf = knn.transform(x_test)
```

```python
x_train_trf = pd.DataFrame(x_train_trf,columns=x_train.columns)
x_test_trf = pd.DataFrame(x_test_trf , columns=x_test.columns)
```
